```python
# This script vectorize reviews for restaurants
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import json
import numpy as np
import pandas as pd
```

```python
restaurant_id = []
restaurant_review = []

with open ("../parsed.dataset/restaurant.keywords.json", 'r') as f:
    restaurants = json.load(f)
    for k,v in restaurants.items():
        restaurant_id.append(k)
        restaurant_review.append(v)
```

```python
assert len(restaurant_id) == len(restaurant_review) == 59371
```

```python
vectorizer = TfidfVectorizer(input=restaurant_review, lowercase=True, min_df=300, ngram_range=(1,1))
X = vectorizer.fit_transform(restaurant_review)
```

```python
matrix = X.toarray()
```

```python
matrix.shape
```

```python
tf_idf = pd.DataFrame(matrix, columns = vectorizer.get_feature_names())
```

```python
tf_idf.index = restaurant_id
tf_idf.index.name = "restaurant_id"
```

```python
tf_idf.head()
```

```python
tf_idf.to_csv("../../tdidf_vectors.tsv", sep = ",", index = True)
```
